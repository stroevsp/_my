Имеем: State, Phone_number, International_calls, Voice_mail, Care --- категориальные переменные (тип `object`), остальные переменные --- количественные.    

Находим частоты значений исследуемых переменных с помощью метода `.value_counts()`. При необходимости редкие значения переменной объединяются в одно значение  

```python
set_columns=[c for c in df.columns] 
for c in set_columns:
    print(df[c].value_counts())
```

Исследуем данные переменные на наличие пропусков. Для этого воспользуемся методом `.value_counts()` со значением параметра `dropna=False`

```python
for c in set_columns:
    print(df[c].value_counts(dropna=False))
```

Вывод общей информации о наличие пропусков во всех переменных можно осуществить, воспользовавшись цепочкой методов `.isnull()` и `.sum()`

```python
df.isnull().sum()
```

Видим наличие пропусков в исследуемых данных

**Заполнение пропусков количественных переменных**  
Пропуски в количественных переменных заполняются значениями вычисленных статистик: среднее, мода. Пропуски также можно заменять константами. 
Для замены пропущенных значений воспользуемся методом `.fillna()` с параметром `inplace` равным `True` для выполнения замены на месте.
Нижеследующий код заполяет пропуски по переменным, перечисленным в списке. 

```python
for i in ['Daytime_minutes_used', 'Number_of_calls_during_the_day', 'Monthly fee', 'Using_night_minutes', 'Number_of_night_calls', 'Nightly_subscription_fee', 'Subscription_fee_for_international_calls']:
    if i in df.columns:
        df[i].fillna(df[i].mean(), inplace=True)
```

ДОПОЛНИТЕЛЬНАЯ ИНФОРМАЦИЯ  
Если необходимо заполнить пробелы по одному признаку, то   
    `df['признак'].fillna(df['признак'].mean(), inplace=True)`

```python
df.isnull().sum()
```

Пропусков в количественных переменных нет.

В данном случае пропущенные значения замещены выборочным средним `.mean()`. Для замены пропущенных значений медианой следует воспользоваться `.median()`.  
Если же пропущенное значение `принака` необходимо заменить `константой`, то  
`df['признак'].fillna(константа, inplace=True)`
